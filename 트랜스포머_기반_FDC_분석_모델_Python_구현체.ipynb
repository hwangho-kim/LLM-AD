```
# ==============================================================================
# 패키지 설치 (Package Installation)
# ==============================================================================
# 아래 명령어를 사용하여 이 스크립트 실행에 필요한 주요 라이브러리들을 설치합니다.
# In a command line or terminal, run the following command:
# pip install pandas numpy tensorflow scikit-learn shap matplotlib seaborn koreanize-matplotlib

# ==============================================================================
# 라이브러리 임포트 (Library Imports)
# ==============================================================================
# 데이터 처리, 모델링, 시각화에 필요한 라이브러리들을 임포트합니다.
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LayerNormalization, MultiHeadAttention, Dropout, GlobalAveragePooling1D, Conv1D, Add
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# Matplotlib에서 한글 폰트가 깨지지 않도록 설정
import koreanize_matplotlib

# 경고 메시지 무시
warnings.filterwarnings('ignore')

# ==============================================================================
# 1단계: 샘플 데이터 생성 (Sample Data Generation)
# ==============================================================================
# (이전 코드와 동일)
def generate_wafer_data(num_wafers=500, timesteps=100):
    """
    하나의 공정 스텝에 대한 웨이퍼 데이터를 생성합니다.
    - FDC: 4개의 센서(압력, RF파워, 가스, 온도)에 대한 시계열 데이터
    - MES: 각 웨이퍼의 공정 조건 및 최종 수율 정보
    """
    print(f"{num_wafers}개 웨이퍼에 대한 샘플 데이터 생성을 시작합니다...")
    wafers = []
    for i in range(num_wafers):
        wafer_id = f'WAF_{i+1:04d}'
        equipment_id = np.random.choice(['Etcher_A', 'Etcher_B', 'Etcher_C'])
        recipe_id = np.random.choice(['Recipe_1', 'Recipe_2'])
        wafers.append([wafer_id, equipment_id, recipe_id])
    mes_df = pd.DataFrame(wafers, columns=['wafer_id', 'equipment_id', 'recipe_id'])

    fdc_data = {}
    anomaly_types = []
    final_yields = []
    for i, row in mes_df.iterrows():
        wafer_id = row['wafer_id']
        pressure_mean, rf_power_mean, gas_flow_mean, temp_mean = 100, 500, 200, 80
        pressure = np.random.normal(pressure_mean, 1, timesteps)
        rf_power = np.random.normal(rf_power_mean, 5, timesteps)
        gas_flow = np.random.normal(gas_flow_mean, 2, timesteps)
        temperature = np.random.normal(temp_mean, 0.5, timesteps)
        anomaly_type, final_yield = 'Normal', 1

        if np.random.rand() < 0.2:
            final_yield = 0
            anomaly_choice = np.random.choice(['Spike', 'Variance', 'Drift'])
            anomaly_type = anomaly_choice
            if anomaly_choice == 'Spike':
                spike_sensor = np.random.choice(['pressure', 'rf_power'])
                spike_idx = np.random.randint(20, 80)
                if spike_sensor == 'pressure': pressure[spike_idx] += pressure_mean * 0.1 * np.sign(np.random.randn())
                else: rf_power[spike_idx] += rf_power_mean * 0.05 * np.sign(np.random.randn())
            elif anomaly_choice == 'Variance':
                var_sensor = np.random.choice(['gas_flow', 'temperature'])
                if var_sensor == 'gas_flow': gas_flow = np.random.normal(gas_flow_mean, 6, timesteps)
                else: temperature = np.random.normal(temp_mean, 1.5, timesteps)
            elif anomaly_choice == 'Drift':
                drift_sensor = np.random.choice(['pressure', 'rf_power'])
                drift_amount = np.linspace(0, (np.random.rand() - 0.5) * 10, timesteps)
                if drift_sensor == 'pressure': pressure += drift_amount
                else: rf_power += drift_amount * 5

        fdc_data[wafer_id] = {'pressure': pressure, 'rf_power': rf_power, 'gas_flow': gas_flow, 'temperature': temperature}
        anomaly_types.append(anomaly_type)
        final_yields.append(final_yield)

    mes_df['anomaly_type'] = anomaly_types
    mes_df['final_yield'] = final_yields
    print("샘플 데이터 생성이 완료되었습니다.")
    return mes_df, fdc_data

mes_df, fdc_data = generate_wafer_data(num_wafers=500, timesteps=100)
print("\n[생성된 MES 데이터 샘플 (상위 5개)]")
print(mes_df.head())
print(f"\n[정상/이상 데이터 분포]\n{mes_df['anomaly_type'].value_counts()}")

fig, axes = plt.subplots(4, 4, figsize=(20, 16), constrained_layout=True)
fig.suptitle('FDC 시계열 데이터 샘플 (정상 및 유형별 이상)', fontsize=20)
for i, anomaly in enumerate(['Normal', 'Spike', 'Variance', 'Drift']):
    sample_wafer_id = mes_df[mes_df['anomaly_type'] == anomaly]['wafer_id'].iloc[0]
    axes[0, i].plot(fdc_data[sample_wafer_id]['pressure']); axes[0, i].set_title(f'유형: {anomaly}\n(Pressure)', fontsize=12)
    axes[1, i].plot(fdc_data[sample_wafer_id]['rf_power']); axes[1, i].set_title('RF Power', fontsize=12)
    axes[2, i].plot(fdc_data[sample_wafer_id]['gas_flow']); axes[2, i].set_title('Gas Flow', fontsize=12)
    axes[3, i].plot(fdc_data[sample_wafer_id]['temperature']); axes[3, i].set_title('Temperature', fontsize=12)
plt.show()

# ==============================================================================
# 2단계: 트랜스포머 모델을 위한 데이터 준비
# ==============================================================================
# 원본 FDC 시계열 데이터를 모델의 입력 형태 (samples, timesteps, features)로 변환하고 스케일링합니다.
print("\n트랜스포머 모델을 위한 데이터 준비를 시작합니다...")
# FDC 데이터를 3D Numpy 배열로 변환
sensor_cols = ['pressure', 'rf_power', 'gas_flow', 'temperature']
data_list = []
for wafer_id in mes_df['wafer_id']:
    wafer_data = [fdc_data[wafer_id][sensor] for sensor in sensor_cols]
    data_list.append(np.array(wafer_data).T) # (timesteps, features) 형태로 변환

X_ts = np.array(data_list) # (samples, timesteps, features)

# 데이터 스케일링
# 각 센서(feature)별로 독립적인 스케일러를 적용합니다.
scalers = {}
X_ts_scaled = np.zeros_like(X_ts)
for i in range(X_ts.shape[2]):
    scalers[i] = StandardScaler()
    X_ts_scaled[:, :, i] = scalers[i].fit_transform(X_ts[:, :, i])

print("데이터 준비가 완료되었습니다.")
print(f"최종 데이터 형태: {X_ts_scaled.shape}")

# 정상/이상 데이터 분리
y = mes_df['final_yield'].values
X_normal_scaled = X_ts_scaled[y == 1]
X_anomaly_scaled = X_ts_scaled[y == 0]

# 학습 및 검증 데이터 분리
X_train, X_val = train_test_split(X_normal_scaled, test_size=0.2, random_state=42)
print(f"학습 데이터 형태: {X_train.shape}")
print(f"검증 데이터 형태: {X_val.shape}")

# ==============================================================================
# 3단계: 트랜스포머 기반 이상 탐지 모델
# ==============================================================================
# 트랜스포머 인코더-디코더 구조를 사용하여 시계열 데이터의 정상 패턴을 학습합니다.

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    """트랜스포머 인코더 블록"""
    x = LayerNormalization(epsilon=1e-6)(inputs)
    x = MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(x, x)
    x = Dropout(dropout)(x)
    res = Add()([x, inputs])

    x = LayerNormalization(epsilon=1e-6)(res)
    x = Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(x)
    x = Dropout(dropout)(x)
    x = Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    return Add()([x, res])

def build_transformer_autoencoder(input_shape, head_size, num_heads, ff_dim, num_transformer_blocks, dropout=0):
    """트랜스포머 오토인코더 모델 구축"""
    inputs = Input(shape=input_shape)
    x = inputs
    # 인코더 부분
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    # 디코더 부분 (여기서는 인코더와 대칭적인 구조로 간단히 구현)
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    # 최종 출력 레이어
    outputs = Dense(input_shape[-1])(x)

    model = Model(inputs, outputs, name="transformer_autoencoder")
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss="mae")
    return model

# 모델 파라미터 설정
input_shape = X_train.shape[1:]
model = build_transformer_autoencoder(
    input_shape,
    head_size=128,
    num_heads=4,
    ff_dim=4,
    num_transformer_blocks=4,
    dropout=0.1,
)
model.summary()

# --- 모델 학습 ---
print("\n트랜스포머 모델 학습을 시작합니다...")
history = model.fit(
    X_train, X_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_val, X_val),
    callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, mode="min")],
    verbose=1
).history

# --- 학습 손실 시각화 ---
plt.figure(figsize=(10, 6))
plt.plot(history['loss'], label='훈련 손실 (Train Loss)')
plt.plot(history['val_loss'], label='검증 손실 (Validation Loss)')
plt.title('트랜스포머 모델 손실', fontsize=16)
plt.xlabel('에포크 (Epoch)', fontsize=12)
plt.ylabel('손실 (Loss)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# --- 이상 점수 계산 및 임계치 설정 ---
X_pred = model.predict(X_ts_scaled)
mae = np.mean(np.abs(X_ts_scaled - X_pred), axis=(1, 2))
mes_df['reconstruction_error'] = mae

train_mae = np.mean(np.abs(X_train - model.predict(X_train)), axis=(1, 2))
threshold = np.quantile(train_mae, 0.99)
print(f"\n계산된 이상 탐지 임계치: {threshold:.4f}")

# --- 복원 오류 분포 시각화 ---
plt.figure(figsize=(10, 6))
sns.histplot(mes_df[mes_df['final_yield'] == 1]['reconstruction_error'], color="blue", label="정상", kde=True, stat="density", linewidth=0)
sns.histplot(mes_df[mes_df['final_yield'] == 0]['reconstruction_error'], color="red", label="이상", kde=True, stat="density", linewidth=0)
plt.axvline(threshold, color='green', linestyle='--', label=f'임계치: {threshold:.4f}')
plt.title('정상/이상 데이터의 복원 오류 분포 (트랜스포머)', fontsize=16)
plt.xlabel('복원 오류 (Reconstruction Error)', fontsize=12)
plt.ylabel('밀도 (Density)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# --- 모델 평가 ---
y_true = mes_df['final_yield'] == 0  # 0(불량)이면 True(이상)
y_pred = mes_df['reconstruction_error'] > threshold

print("\n[이상 탐지 모델 평가 결과 (트랜스포머)]")
print(classification_report(y_true, y_pred, target_names=['정상 (Normal)', '이상 (Anomaly)']))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['정상 (Normal)', '이상 (Anomaly)'], yticklabels=['정상 (Normal)', '이상 (Anomaly)'])
plt.title('혼동 행렬 (Confusion Matrix) - 트랜스포머', fontsize=16)
plt.ylabel('실제 값 (Actual)', fontsize=12)
plt.xlabel('예측 값 (Predicted)', fontsize=12)
plt.show()


# ==============================================================================
# 4단계: 원인 변수 설명 (Explainable AI)
# ==============================================================================
# 주석: 트랜스포머와 같은 복잡한 시계열 모델의 재구성 오류를 설명하는 것은
# SHAP과 같은 도구를 적용하기에 매우 복잡하고 계산 비용이 많이 듭니다.
# 실제 시스템에서는 이상이 탐지된 후, 해당 시계열의 통계적 특징(예: 평균, 표준편차)을
# 계산하여 어떤 특징이 비정상적인지를 분석하는 하이브리드 접근법을 사용할 수 있습니다.
# 이 예제에서는 탐지 모델 교체에 집중하기 위해 XAI 부분은 생략합니다.
print("\n[4단계: 원인 분석] - 트랜스포머 기반 탐지 모델 구현에 집중하기 위해 이 단계는 생략되었습니다.")
```
